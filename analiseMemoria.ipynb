```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
#ler csv
df_flutter = pd.read_csv('mem_experimento_flutter.csv')
df_rn = pd.read_csv('mem_experimento_rn.csv')
```

```python
# Agrupar por tempo e calcular a média do consumo de memória (uso médio de memória por tempo, considerando todos os experimentos)
df_flutter_media = df_flutter.groupby("tempo")["consumo_memoria"].mean().reset_index()
df_flutter_media
```

```python
df_rn_media = df_rn.groupby("tempo")["consumo_memoria"].mean().reset_index()
df_rn_media
```

```python
'''# Calcular a EMA/MME (suavização exponencial através da Média Móvel Exponencial/Exponential Moving Average)
df_media["consumo_memoria_MME"] = df_media["consumo_memoria"].ewm(alpha=0.6, adjust=False).mean()
df_media'''
```

```python
# Encontrar o pico médio (considerando todos os experimentos)

# Pico - Flutter
pico_idx_flutter = df_flutter_media["consumo_memoria"].idxmax()
pico_tempo_flutter = df_flutter_media.loc[pico_idx_flutter, "tempo"]
pico_mem_flutter = df_flutter_media.loc[pico_idx_flutter, "consumo_memoria"]

# Pico - React Native
pico_idx_rn = df_rn_media["consumo_memoria"].idxmax()
pico_tempo_rn = df_rn_media.loc[pico_idx_rn, "tempo"]
pico_mem_rn = df_rn_media.loc[pico_idx_rn, "consumo_memoria"]

# Valor mínimo médio - Flutter
min_idx_flutter = df_flutter_media["consumo_memoria"].idxmin()
min_tempo_flutter = df_flutter_media.loc[min_idx_flutter, "tempo"]
min_mem_flutter = df_flutter_media.loc[min_idx_flutter, "consumo_memoria"]

# Valor mínimo médio - React Native
min_idx_rn = df_rn_media["consumo_memoria"].idxmin()
min_tempo_rn = df_rn_media.loc[min_idx_rn, "tempo"]
min_mem_rn = df_rn_media.loc[min_idx_rn, "consumo_memoria"]
```

```python
# Plot

# Criar figura
fig = go.Figure()

# Curva Flutter
fig.add_trace(go.Scatter(
    x=df_flutter_media["tempo"],
    y=df_flutter_media["consumo_memoria"],
    mode='lines',
    name='Consumo de memória em Flutter',
    line=dict(color='orange', width=3),
    opacity=0.5
))


'''# Linha MME
fig.add_trace(go.Scatter(
    x=df_media["tempo"],
    y=df_media["consumo_memoria_MME"],
    mode='lines',
    name='MME do consumo de memória RAM (suavização)',
    line=dict(color='orange', width=3)
))'''

# Curva React Native
fig.add_trace(go.Scatter(
    x=df_rn_media["tempo"],
    y=df_rn_media["consumo_memoria"],
    mode='lines',
    name='Consumo de memória em React Native',
    line=dict(color='blue', width=3),
    opacity=0.5
))

# Marcar pico Flutter
fig.add_trace(go.Scatter(
    x=[pico_tempo_flutter],
    y=[pico_mem_flutter],
    mode='markers+text',
    name='Pico',
    marker=dict(color='red', size=10),
    text=[f"{pico_mem_flutter:.2f} MB".replace('.',',')],
    textposition="top center"
))

# Marcar pico React Native (sem duplicar legenda)
fig.add_trace(go.Scatter(
    x=[pico_tempo_rn],
    y=[pico_mem_rn],
    mode='markers+text',
    name='Pico',
    showlegend=False,
    marker=dict(color='red', size=10),
    text=[f"{pico_mem_rn:.2f} MB".replace('.',',')],
    textposition="top center"
))

# Marcar mínimo Flutter
fig.add_trace(go.Scatter(
    x=[min_tempo_flutter],
    y=[min_mem_flutter],
    mode='markers+text',
    name='Mínimo',
    marker=dict(color='green', size=10),
    text=[f"{min_mem_flutter:.2f} MB".replace('.',',')],
    textposition="bottom center"
))

# Marcar mínimo React Native (sem duplicar legenda)
fig.add_trace(go.Scatter(
    x=[min_tempo_rn],
    y=[min_mem_rn],
    mode='markers+text',
    name='Mínimo',
    showlegend=False,
    marker=dict(color='green', size=10),
    text=[f"{min_mem_rn:.2f} MB".replace('.',',')],
    textposition="bottom center"
))

# Layout completo
fig.update_layout(
    width=900,
    height=600,
    xaxis=dict(
        title='Tempo (s)',
        title_font=dict(family='Arial', size=12, color='black',),
        tickfont=dict(family='Arial', size=12, color='black'),
        showgrid=True,
        gridwidth=1.5,
        gridcolor='#e5ecf6'
    ),
    yaxis=dict(
        title='Consumo de memória (MB)',
        title_font=dict(family='Arial', size=12, color='black'),
        tickfont=dict(family='Arial', size=12, color='black'),
        showgrid=True,
        gridwidth=1.5,
        gridcolor='#e5ecf6'
    ),
    legend=dict(
        x=1, y=1.23, xanchor='right', yanchor='top',
        font=dict(family='Arial', size=12, color='black')
    ),
    template='plotly_white'
)


fig.show()
```
