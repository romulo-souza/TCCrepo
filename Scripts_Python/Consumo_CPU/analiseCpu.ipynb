```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

#Resultados React Native x Flutter. Alterar CSVs para cada frequência da atualização de elementos utilizada


```python
#ler csv (um csv para cada frequência de atualização usada)
df_flutter = pd.read_csv('cpu_experimento_flutter_400ms.csv')
df_rn = pd.read_csv('cpu_experimento_rn_400ms.csv')
```

##Agrupar por tempo e calcular a média do consumo de CPU (uso médio de CPU por tempo, considerando todos os experimentos)

```python
df_flutter_media = df_flutter.groupby("tempo")["consumo_cpu"].mean().reset_index()
df_flutter_media
```

```python
df_rn_media = df_rn.groupby("tempo")["consumo_cpu"].mean().reset_index()
df_rn_media
```

## Pico e mínimo de consumo dos valores médios

```python
# Pico - Flutter
pico_idx_flutter = df_flutter_media["consumo_cpu"].idxmax()
pico_tempo_flutter = df_flutter_media.loc[pico_idx_flutter, "tempo"]
pico_cpu_flutter = df_flutter_media.loc[pico_idx_flutter, "consumo_cpu"]

# Pico - React Native
pico_idx_rn = df_rn_media["consumo_cpu"].idxmax()
pico_tempo_rn = df_rn_media.loc[pico_idx_rn, "tempo"]
pico_cpu_rn = df_rn_media.loc[pico_idx_rn, "consumo_cpu"]

# Valor mínimo médio - Flutter
min_idx_flutter = df_flutter_media["consumo_cpu"].idxmin()
min_tempo_flutter = df_flutter_media.loc[min_idx_flutter, "tempo"]
min_cpu_flutter = df_flutter_media.loc[min_idx_flutter, "consumo_cpu"]

# Valor mínimo médio - React Native
min_idx_rn = df_rn_media["consumo_cpu"].idxmin()
min_tempo_rn = df_rn_media.loc[min_idx_rn, "tempo"]
min_cpu_rn = df_rn_media.loc[min_idx_rn, "consumo_cpu"]

```

##Gráficos

```python
# Plot

# Criar figura
fig = go.Figure()

# Curva Flutter
fig.add_trace(go.Scatter(
    x=df_flutter_media["tempo"],
    y=df_flutter_media["consumo_cpu"],
    mode='lines',
    name='Consumo de CPU em Flutter',
    line=dict(color='orange', width=3),
    opacity=0.5
))

# Curva React Native
fig.add_trace(go.Scatter(
    x=df_rn_media["tempo"],
    y=df_rn_media["consumo_cpu"],
    mode='lines',
    name='Consumo de CPU em React Native',
    line=dict(color='blue', width=3),
    opacity=0.5
))

# Marcar pico Flutter
fig.add_trace(go.Scatter(
    x=[pico_tempo_flutter],
    y=[pico_cpu_flutter],
    mode='markers+text',
    name='Pico',
    marker=dict(color='red', size=10),
    text=[f"{pico_cpu_flutter:.2f}%".replace('.',',')],
    textposition="top center"
))

# Marcar pico React Native (sem duplicar legenda)
fig.add_trace(go.Scatter(
    x=[pico_tempo_rn],
    y=[pico_cpu_rn],
    mode='markers+text',
    name='Pico',
    showlegend=False,
    marker=dict(color='red', size=10),
    text=[f"{pico_cpu_rn:.2f}%".replace('.',',')],
    textposition="top center"
))

# Marcar mínimo Flutter
fig.add_trace(go.Scatter(
    x=[min_tempo_flutter],
    y=[min_cpu_flutter],
    mode='markers+text',
    name='Mínimo',
    marker=dict(color='green', size=10),
    text=[f"{min_cpu_flutter:.2f}%".replace('.',',')],
    textposition="bottom center"
))

# Marcar mínimo React Native (sem duplicar legenda)
fig.add_trace(go.Scatter(
    x=[min_tempo_rn],
    y=[min_cpu_rn],
    mode='markers+text',
    name='Mínimo médio',
    showlegend=False,
    marker=dict(color='green', size=10),
    text=[f"{min_cpu_rn:.2f}%".replace('.',',')],
    textposition="bottom center"
))

# Layout completo
fig.update_layout(
    width=950,
    height=600,

    xaxis=dict(
        title='Tempo (s)',
        title_font=dict(family='Arial', size=12, color='black'),
        tickfont=dict(family='Arial', size=12, color='black'),
        showgrid=True,
        gridwidth=1.5,
        gridcolor='#e5ecf6'
    ),
    yaxis=dict(
        title='Consumo de CPU (%)',
        title_font=dict(family='Arial', size=12, color='black'),
        tickfont=dict(family='Arial', size=12, color='black'),
        showgrid=True,
        gridwidth=1.5,
        gridcolor='#e5ecf6'
    ),
    legend=dict(
        x=0.98, y=1.15, xanchor='right', yanchor='top',
        font=dict(family='Arial', size=12, color='black')
    ),
    template='plotly_white'
)

fig.show()
```
