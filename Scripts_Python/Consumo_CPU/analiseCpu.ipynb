```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
#ler csv (um csv para cada frequencia de tempo usada)
df = pd.read_csv('cpu_experimentos.csv')
```

```python
df
```

```python
df.info()
```

```python
# Agrupar por tempo e calcular a mÃ©dia do consumo de CPU (uso mÃ©dio de CPU por tempo, considerando todos os experimentos)
df_media = df.groupby("tempo")["consumo_cpu"].mean().reset_index()
df_media
```

```python
# Calcular a EMA (suavizaÃ§Ã£o exponencial atravÃ©s da MÃ©dia MÃ³vel Exponencial/Exponential Moving Average)
df_media["consumo_cpu_MME"] = df_media["consumo_cpu"].ewm(alpha=0.6, adjust=False).mean()
df_media
```

```python
# Encontrar o pico mÃ©dio (considerando todos experimentos)
pico_idx = df_media["consumo_cpu"].idxmax()
pico_tempo = df_media.loc[pico_idx, "tempo"]
pico_cpu = df_media.loc[pico_idx, "consumo_cpu"]
```

```python
# Plot

# Criar figura
fig = go.Figure()

# Linha pontilhada para dados reais
fig.add_trace(go.Scatter(
    x=df_media["tempo"],
    y=df_media["consumo_cpu"],
    mode='lines',
    name='Consumo de CPU (dados brutos)',
    line=dict(dash='dot', color='blue', width=2),
    opacity=0.4
))

# Linha MME
fig.add_trace(go.Scatter(
    x=df_media["tempo"],
    y=df_media["consumo_cpu_MME"],
    mode='lines',
    name='MME do consumo de CPU (suavizaÃ§Ã£o)',
    line=dict(color='orange', width=3)
))

# Marcar pico
fig.add_trace(go.Scatter(
    x=[pico_tempo],
    y=[pico_cpu],
    mode='markers+text',
    name='Pico',
    marker=dict(color='red', size=10),
    text=[f"Pico: {pico_cpu:.2f}%"],
    textposition="top left"
))

# Layout completo com fontes Arial 12 e cor preta
fig.update_layout(
    width=950,
    height=550,
    title=dict(
        text='Consumo mÃ©dio de CPU',
        font=dict(family='Arial', size=12, color='black', weight='bold'),
        xanchor='center',
        x=0.5,
    ),
    xaxis=dict(
        title='Tempo (s)',
        title_font=dict(family='Arial', size=12, color='black'),
        tickfont=dict(family='Arial', size=12, color='black'),
        showgrid=True,
        gridwidth=1.5,
        gridcolor='#e5ecf6'
    ),
    yaxis=dict(
        title='Consumo de CPU (%)',
        title_font=dict(family='Arial', size=12, color='black'),
        tickfont=dict(family='Arial', size=12, color='black'),
        showgrid=True,
        gridwidth=1.5,
        gridcolor='#e5ecf6'
    ),
    legend=dict(
        x=1.12, y=1.25, xanchor='right', yanchor='top',
        font=dict(family='Arial', size=12, color='black')
    ),
    template='plotly_white'
)

fig.show()
```
