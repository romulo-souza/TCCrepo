```python
import pandas as pd
import plotly.express as px
```

# Resultados Flutter

```python
df_flutter = pd.read_csv('tempoApiFlutter.csv')
```

```python
df_flutter.groupby('tamanho_json')['tempo_resposta'].describe().round(2)
```

# Resultados React Native

```python
df_rn = pd.read_csv('tempoApiRN.csv')
```

```python
df_rn.groupby('tamanho_json')['tempo_resposta'].describe().round(2)
```

# Boxplot

```python
# identificar frameworks
df_flutter["framework"] = "Flutter"
df_rn["framework"] = "React Native"

# concatenar em um Ãºnico df
df_all = pd.concat([df_flutter, df_rn], ignore_index=True)

# Converter os tamanhos para string (categorias para facilitar no agrupamento do boxplot (eixo x))
df_all["tamanho_json"] = df_all["tamanho_json"].astype(str)

df_all.info()
```

```python
# boxplot
# Criar boxplot
fig = px.box(
    df_all,
    x="tamanho_json",
    y="tempo_resposta",
    color="framework",
    title="Boxplots dos tempos de resposta por framework e tamanho do JSON",
    color_discrete_map={"React Native": "blue", "Flutter": "orange"},
)

fig.update_layout(
    height=850,
    width=650,
    legend_title = "",
    yaxis_title="Tempo de resposta (ms)",
    xaxis_title="Tamanho do JSON (KB)",
    boxmode="group",  # agrupar boxplots do Flutter e React Native lado a lado
    font=dict(
        family="Arial",
        size=12,
        color="black"
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.0,
        xanchor="center",
        x=0.5
    ),
    xaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    ),
    yaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    )
)

fig.show()
```
