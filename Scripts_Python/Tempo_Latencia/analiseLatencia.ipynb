```python
import pandas as pd
import plotly.express as px
```

```python
#ler csv
df_flutter = pd.read_csv('latenciaEntradaFlutter.csv')
df_rn = pd.read_csv('latenciaEntradaRN.csv')
```

# Resultados Flutter

```python
df_flutter.describe().round(2)
```

# Resultados React Native

```python
df_rn.describe().round(2)
```

#Boxplot

```python
# Unir os 2 Dataframes

# Adiciona uma coluna para identificar a origem
df_rn['framework'] = 'React Native'
df_flutter['framework'] = 'Flutter'

df_final = pd.concat([df_rn, df_flutter], ignore_index=True)
df_final
```

```python
# Grfico
# Grfico
fig = px.box(
    df_final,
    x='framework',
    y='latencia_entrada',
    color='framework',
    title='', #Boxplot das latncias de entrada por framework
    color_discrete_map={'React Native': 'blue', 'Flutter': 'orange'}
)

fig.update_layout(
    yaxis_title='Latncia (ms)',
    xaxis_title='Framework',
    legend_title='',
    width=600,
    height=600,
    font=dict(
        family="Arial",
        size=12,
        color="black"
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="center",
        x=0.5
    ),
    xaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    ),
    yaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    )
)

fig.show()
```
