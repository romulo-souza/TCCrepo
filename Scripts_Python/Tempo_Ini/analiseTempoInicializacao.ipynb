```python
import pandas as pd
```

```python
#ler csv
df_flutter = pd.read_csv('resultados_flutter.csv')
df_rn = pd.read_csv('resultados_rn.csv')
```

# Resultados Flutter

```python
df_flutter
```

```python
df_flutter['tipo_inicializacao'].value_counts()
```

```python
# Agrupar por tipo de inicialização e calcular o tempo médio de cada um
df_flutter_media = df_flutter.groupby('tipo_inicializacao')['tempo'].mean().round(2).sort_values(ascending=False).reset_index()
df_flutter_media
```

```python
df_rn['tipo_inicializacao'].value_counts()
```

```python
df_rn_media = df_rn.groupby('tipo_inicializacao')['tempo'].mean().round(2).sort_values(ascending=False).reset_index()
df_rn_media
```
