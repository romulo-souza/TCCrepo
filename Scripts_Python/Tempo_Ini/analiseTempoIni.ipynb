```python
import pandas as pd
import plotly.express as px
```

## Gráfico boxplot

```python
# identificar frameworks
df_flutter["framework"] = "Flutter"
df_rn["framework"] = "React Native"

# concatenar em um único df
df_all = pd.concat([df_flutter, df_rn], ignore_index=True)

# deixar a primeira letra maiúscula para o tipo de inicialização
df_all['tipo_inicializacao'] = df_all["tipo_inicializacao"].str.capitalize()
df_all
```

```python
# boxplot
# Criar boxplot
fig = px.box(
    df_all,
    x="tipo_inicializacao",
    y="tempo",
    color="framework",
    title="Boxplot do tempo de inicialização por tipo e framework",
    color_discrete_map={"React Native": "blue", "Flutter": "orange"},
)

fig.update_layout(
    height=850,
    width=650,
    legend_title = "",
    yaxis_title="Tempo (ms)",
    xaxis_title="Tipo de inicialização",
    boxmode="group",  # agrupar boxplots do Flutter e React Native lado a lado
    font=dict(
        family="Arial",
        size=12,
        color="black"
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.0,
        xanchor="center",
        x=0.5
    ),
    xaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    ),
    yaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    )
)

fig.show()
```
