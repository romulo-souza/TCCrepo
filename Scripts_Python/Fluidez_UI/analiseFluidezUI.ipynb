## Boxplot


```python
# Unir os 2 Dataframes

# Adiciona uma coluna para identificar a origem
df_rn['framework'] = 'React Native'
df_flutter['framework'] = 'Flutter'

df_final = pd.concat([df_rn, df_flutter])
```

```python
df_final
```

```python
fig = px.box(
    df_final,
    x='framework',
    y='janky_frames',
    color='framework',
    title='Boxplot de Janky Frames por Framework',
    color_discrete_map={'React Native': 'blue', 'Flutter': 'orange'}
)

fig.update_layout(
    yaxis_title='Janky Frames',
    xaxis_title='Framework',
    legend_title='',
    width=600,
    font=dict(
        family="Arial",
        size=12,
        color="black"
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.05,
        xanchor="center",
        x=0.5
    ),
    xaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    ),
    yaxis=dict(
        showgrid=True,
        gridwidth=1,
        gridcolor='rgba(0,0,0,0.1)'
    )
)

fig.show()
```
